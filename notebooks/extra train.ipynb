```python
df = pandas.read_table("/Users/tim/Dropbox/sinai/projects/phip-seq-ml/201711/data/twin-virscan-peptide-training.tsv")
df
```

```python
sample_cols = list(df.columns)[3:]
sample_cols
```

```python
len(sample_cols), df.uniprot_accession.nunique(), df.id.nunique()
```

```python
stacked_df = df.set_index(["id", "uniprot_accession", "sequence_aa"])[sample_cols].stack().reset_index()
stacked_df.columns = list(stacked_df.columns)[:3] + ["specimen", "value"]

max_uniprot = stacked_df.groupby("uniprot_accession").value.max()
max_id = stacked_df.groupby("id").value.max()
num_ids_per_uniprot = df.groupby("uniprot_accession").id.nunique()

stacked_df["num_sequences"] = stacked_df.uniprot_accession.map(num_ids_per_uniprot)
stacked_df["max_uniprot"] = stacked_df.uniprot_accession.map(max_uniprot)
stacked_df["max_id"] = stacked_df.id.map(max_id)
stacked_df["ratio"] = stacked_df.max_id / stacked_df.max_uniprot
stacked_df
```

```python
grouped_df = stacked_df.groupby(
    ["id", "uniprot_accession", "sequence_aa", "num_sequences", "max_uniprot", "max_id"]).ratio.max().reset_index()
grouped_df
```

```python
threshold = 100
selected_df = grouped_df.loc[grouped_df.max_uniprot >= threshold].reset_index().copy()
del selected_df['index']
selected_df
```
